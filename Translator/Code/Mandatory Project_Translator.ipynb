```python
# Import the Pandas and GoogleTrans modules
import pandas as pd
from googletrans import Translator

# Load the workbook and select the sheet
df = pd.read_csv(r'/content/df_arabica_clean.csv')

# Initialize the Google Translator
translator = Translator()

# Define the column to translate (e.g., column A)
column_to_translate = ['Lot Number', 'Region']

# Function to translate the Simplified Chinese to English
def translate_ch_text(text):
    try:
        translated = translator.translate(text, src='zh-cn', dest='en').text
        return translated
    except Exception as e:
        return text

# Function to translate the any language to English
def translate_text(text):
    try:
        translated = translator.translate(text, src='auto', dest='en').text
        return translated
    except Exception as e:
        return text

# Create an empty DataFrame to store new columns
translated_df = pd.DataFrame()

# Translating each row in the column using map function and adding the translated column
translated_df[column_to_translate] = df[column_to_translate].map(translate_ch_text)
translated_df['In-Country Partner'] = df['In-Country Partner'].map(translate_text)
```

```python
# Export the data to CSV file
translated_df.to_csv('Translated_data.csv', index=False)
```
